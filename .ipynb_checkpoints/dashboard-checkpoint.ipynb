## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    avg_housing_units_df = sfo_data["housing_units"].groupby(sfo_data.index).mean()
    
    fig_housing_plot = plt.figure()
    
    housing_units_plot = avg_housing_units_df.plot.bar()
    housing_units_plot.set_xlabel("Year")
    housing_units_plot.set_ylabel("Housing Units")
    housing_units_plot.set_title("Housing Units in San Francisco from 2010 to 2016")
    
    housing_units_std = avg_housing_units_df.std()
    housing_units_min = min(avg_housing_units_df)-housing_units_std 
    housing_units_max = max(avg_housing_units_df)+housing_units_std 
    
    housing_units_plot.set_ylim(housing_units_min, housing_units_max)
    plt.close(fig_housing_plot)
    
    return pn.pane.Matplotlib(fig_housing_plot, tight=True)
    
def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    avg_sqft_and_rent = (sfo_data[["sale_price_sqr_foot", "gross_rent"]].groupby(sfo_data.index).mean())
    fig_gross_plot = plt.figure()
    gross_rent_plot = avg_sqft_and_rent['gross_rent'].plot(color="red")
    gross_rent_plot.set_xlabel('Year')
    gross_rent_plot.set_ylabel("Gross Rent")
    gross_rent_plot.set_title("Average Gross Rent Per Year ")
    plt.close(fig_gross_plot)
    
    return pn.pane.Matplotlib(fig_gross_plot, tight=True)


def average_sales_price():
    """Average Sales Price Per Year."""
    
    avg_sqft_and_rent = (sfo_data[["sale_price_sqr_foot", "gross_rent"]].groupby(sfo_data.index).mean())
    fig_avg_plot = plt.figure()
    avg_sqrft_plot = avg_sqft_and_rent['sale_price_sqr_foot'].plot()
    avg_sqrft_plot.set_xlabel('Year')
    avg_sqrft_plot.set_ylabel("Price/sqft")
    avg_sqrft_plot.set_title("Yearly Average Sale Price Per Square Foot")
    plt.close(fig_avg_plot)

    return pn.pane.Matplotlib(fig_avg_plot, tight=True)

def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    neighborhood_costs = sfo_data.groupby([sfo_data.index, "neighborhood"]).mean()
    neighborhood_costs.reset_index(inplace=True)
    neighborhood_costs.drop(columns=['housing_units'], inplace=True)
    neighborhood_costs.rename(columns={
    "neighborhood": "Neighborhood"}, inplace=True)

    neighborhood_plot = neighborhood_costs.hvplot.line("year","sale_price_sqr_foot", xlabel="Year", ylabel="Avg Sale Price/sqrft", groupby="Neighborhood").opts(
    yformatter=NumeralTickFormatter(format="$0,000.00"))
    
    return neighborhood_plot

def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""

    neighborhood_sqft= sfo_data[['sale_price_sqr_foot', 'neighborhood']]
    neighborhood_sqft_avg = neighborhood_sqft.groupby('neighborhood').mean()
    
    top_10_neighborhoods = neighborhood_sqft_avg.nlargest(10,"sale_price_sqr_foot")
    
    top_10_plot = top_10_neighborhoods.hvplot.bar(
        title="Top 10 Most Expensive Neighborhoods", width=1000,height=400, rot=60, xlabel="Neighborhood", ylabel="Price/sqrft").opts(
    yformatter=NumeralTickFormatter(format="$0,000.00"))
    
    return top_10_plot


def most_expensive_neighborhoods_rent_sales():
    """Comparison of Rent and Sales Prices of Most Expensive Neighborhoods."""   
    
    neighborhood_costs = sfo_data.groupby([sfo_data.index, "neighborhood"]).mean()
    neighborhood_costs.reset_index(inplace=True)
    neighborhood_costs.drop(columns=['housing_units'], inplace=True)
    neighborhood_costs.rename(columns={
    "neighborhood": "Neighborhood"}, inplace=True)
    
    rent_sales_plot = neighborhood_costs.hvplot.bar(
    "year",
    ["gross_rent","sale_price_sqr_foot"],
    title="Top 10 Expensive Neighborhoods in SFO",
    xlabel="Neighborhood",
    ylabel="Gross Rent & Price/sqft",
    groupby='Neighborhood',
    height=400,
    rot=90
    ).opts(yformatter=NumeralTickFormatter(format="$0,000.00"))
    
    return rent_sales_plot
    
    
def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
     

def parallel_categories():
    """Parallel Categories Plot."""



def neighborhood_map():
    """Neighborhood Map."""
    
    px.set_mapbox_access_token(map_box_api)

    coordinates_path = Path("Data/neighborhoods_coordinates.csv")
    df_neighborhood_coordinates = pd.read_csv(coordinates_path)
    
    df_neighborhood_mean = sfo_data.groupby([sfo_data.index, "neighborhood"]).mean()
    df_neighborhood_mean.reset_index(inplace=True)
    df_neighborhood_mean.rename(columns={
    "neighborhood": "Neighborhood"}, inplace=True)
    df_neighborhood_mean.drop(columns=['year', 'Neighborhood'], inplace=True)
    
    neighborhood_joined_data = pd.concat([df_neighborhood_coordinates,df_neighborhood_mean], axis='columns', join='inner')
    
    #return neighborhood_joined_data
    #px.set_mapbox_access_token(map_box_api)

    neighorhood_map = px.scatter_mapbox(
    neighborhood_joined_data,
    lat="Lat",
    lon="Lon",
    size="sale_price_sqr_foot",
    color="gross_rent",
    size_max = 14,
    zoom = 12,
    hover_name="Neighborhood",
    title="Average Sale Price Per Square Foot and Gross Rent in San Francisco",
    )

    plotly_panel = pn.pane.Plotly(neighborhood_map)
    #plotly_panel._updates = True
    
    return plotly_panel

def sunburst():
    """Sunburst Plot."""
    
   
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
title = pn.pane.Markdown(
    """
# Real Estate Analysis of San Francisco from 2010 to 2016
""",
    width=900,
)

summary = pn.pane.Markdown(
    """
Historical pricing data of housing units, sale price per square foot and gross rent in Sanfrancisco from 2010-2016.
You can use the tabs to navigate the different analyisis of the SF housing market. 
"""
)

# Create a tab layout for the dashboard
tabs = pn.Tabs(
    ("Summary", pn.Column(summary)),
    ("Yearly Market Analysis", pn.Row(housing_units_per_year(), average_gross_rent(), average_sales_price())),
    ("Neighborhood Analysis", pn.Column(average_price_by_neighborhood(), top_most_expensive_neighborhoods(), most_expensive_neighborhoods_rent_sales())),
    )
```

```python
# Create the dashboard
dashboard = pn.Column(pn.Row(title), tabs, width=900)
```

## Serve the Panel Dashboard

```python
# Serve the# dashboard
dashboard.servable()
```
